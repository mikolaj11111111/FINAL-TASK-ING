```python
import os
import time
import psutil
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

###############################################################################
#                            FEED-FORWARD NN MODEL                            #
###############################################################################
class FeedForwardNN(nn.Module):
    def __init__(self, input_dim, final_activation='prelu'):
        super(FeedForwardNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.first_activation = nn.LeakyReLU(0.1)

        if final_activation.lower() == 'leaky_relu':
            self.final_activation = nn.LeakyReLU(0.1)
        elif final_activation.lower() == 'prelu':
            self.final_activation = nn.PReLU()
        elif final_activation.lower() == 'gelu':
            self.final_activation = nn.GELU()
        elif final_activation.lower() == 'softplus':
            self.final_activation = nn.Softplus()
        else:
            self.final_activation = nn.ReLU()

        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        x = self.fc1(x)
        x = self.first_activation(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.dropout(x)
        x = self.fc3(x)
        x = self.final_activation(x)
        return x

###############################################################################
#               TRAIN LR & NN FOR A SINGLE CLUSTER (WITH METRICS)             #
###############################################################################
def train_models_on_cluster(
    df_cluster,
    market_label,
    start_yr,
    end_yr,
    cluster_id,
    feature_cols=None,
    target_col="VALUATION_VALUE",
    nn_epochs=2000,       # 2,000 epochs
    nn_patience=500,      # early stopping patience
    nn_models_dir="nn_models"
):
    """
    Trains:
      1) Linear Regression
      2) Neural Network (PyTorch) with early stopping

    Prints metrics for both (MAE, MSE, MAPE, R²).
    Saves the NN model's state_dict to a .pth file,
    and returns a dictionary referencing the LR model, NN file path, etc.
    """
    if feature_cols is None:
        # Example default features
        feature_cols = [
            "PREMISSES_USABLE_AREA",
            "PREMISSES_TECHNICAL_CONDITION",
            "CONSTRUCTION_YEAR","PREMISSES_CHAMBER_NO","BUILDING_TECHNICAL_CONDITION"
        ]

    # If there's insufficient data, skip
    if df_cluster.shape[0] < 5:
        print(f"[WARNING] Cluster {cluster_id} in {market_label} {start_yr}-{end_yr} has <5 rows. Skipping.")
        return None

    # Make sure we have a place to save NN .pth files
    os.makedirs(nn_models_dir, exist_ok=True)

    # ----------------- Train/test split -----------------
    train_df, test_df = train_test_split(df_cluster, test_size=0.3, random_state=42)
    X_train = train_df[feature_cols].values
    y_train = train_df[target_col].values
    X_test  = test_df[feature_cols].values
    y_test  = test_df[target_col].values

    #==========================================================================
    # (1) LINEAR REGRESSION
    #==========================================================================
    lr_model = LinearRegression()
    lr_model.fit(X_train, y_train)

    lr_pred = lr_model.predict(X_test)
    lr_mae  = mean_absolute_error(y_test, lr_pred)
    lr_mse  = mean_squared_error(y_test, lr_pred)
    lr_mape = np.mean(np.abs((y_test - lr_pred) / y_test)) * 100 if np.any(y_test != 0) else 0
    lr_r2   = r2_score(y_test, lr_pred)

    print(f"\n[LR] {market_label} {start_yr}-{end_yr}, Cluster {cluster_id}")
    print(f"   MAE:  {lr_mae:.4f}")
    print(f"   MSE:  {lr_mse:.4f}")
    print(f"   MAPE: {lr_mape:.2f}%")
    print(f"   R²:   {lr_r2:.4f}")

    #==========================================================================
    # (2) NEURAL NETWORK
    #==========================================================================
    #   - scale features with MinMax
    nn_scaler = MinMaxScaler()
    X_train_nn = nn_scaler.fit_transform(X_train)
    X_test_nn  = nn_scaler.transform(X_test)

    X_train_torch = torch.tensor(X_train_nn, dtype=torch.float32)
    y_train_torch = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
    X_test_torch  = torch.tensor(X_test_nn,  dtype=torch.float32)
    y_test_torch  = torch.tensor(y_test,     dtype=torch.float32).view(-1, 1)

    input_dim = X_train_nn.shape[1]
    model = FeedForwardNN(input_dim=input_dim, final_activation="prelu")
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.01)

    best_loss = float("inf")
    best_state = None
    patience_counter = 0

    for epoch in range(nn_epochs):
        model.train()
        optimizer.zero_grad()

        y_pred_torch = model(X_train_torch)
        loss = criterion(y_pred_torch, y_train_torch)
        loss.backward()
        optimizer.step()

        # Early stopping logic
        if loss.item() < best_loss:
            best_loss = loss.item()
            best_state = model.state_dict().copy()
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= nn_patience:
                break

    if best_state is not None:
        model.load_state_dict(best_state)
    model.eval()

    with torch.no_grad():
        nn_out = model(X_test_torch).numpy()

    nn_mae  = mean_absolute_error(y_test, nn_out)
    nn_mse  = mean_squared_error(y_test, nn_out)
    nn_mape = np.mean(np.abs((y_test - nn_out) / y_test)) * 100 if np.any(y_test != 0) else 0
    nn_r2   = r2_score(y_test, nn_out)

    print(f"\n[NN] {market_label} {start_yr}-{end_yr}, Cluster {cluster_id}")
    print(f"   MAE:  {nn_mae:.4f}")
    print(f"   MSE:  {nn_mse:.4f}")
    print(f"   MAPE: {nn_mape:.2f}%")
    print(f"   R²:   {nn_r2:.4f}")

    #==========================================================================
    # Save NN model weights to .pth file
    #==========================================================================
    nn_filename = (f"{market_label}_{start_yr}_{end_yr}_cluster{cluster_id}.pth"
                   .replace(" ", "_").replace("(", "").replace(")", ""))
    nn_path = os.path.join(nn_models_dir, nn_filename)
    torch.save(model.state_dict(), nn_path)

    # Return references needed for inference
    return {
        "lr_model": lr_model,        # scikit-learn LR
        "nn_path": nn_path,          # path to .pth file on disk
        "nn_scaler": nn_scaler,      # MinMaxScaler for the cluster
        "nn_input_dim": input_dim,
        "feature_cols": feature_cols
    }

###############################################################################
#  train_models_for_all_clusters
###############################################################################
def train_models_for_all_clusters(
    clustered_df,
    market_label,
    feature_cols=None,
    nn_epochs=2000,      # 2,000 epochs
    nn_patience=500
):
    """
    Groups the cluster-labeled DataFrame by (Year range, KMeans_Cluster),
    trains LR & NN for each cluster, prints metrics, saves each NN's weights,
    and returns a dictionary keyed by ((start_yr, end_yr), cluster_id).
    """
    if feature_cols is None:
        feature_cols = [
            "PREMISSES_USABLE_AREA",
            "PREMISSES_TECHNICAL_CONDITION",
            "PREMISSES_CHAMBER_NO",
            "BUILDING_TECHNICAL_CONDITION"
            # etc...
        ]

    model_dict = {}

    min_yr = clustered_df["Year"].min()
    max_yr = clustered_df["Year"].max()
    # We'll assume 3-year steps
    all_year_breaks = range(min_yr, max_yr + 1, 3)

    for start_y in all_year_breaks:
        end_y = start_y + 2
        mask = (clustered_df["Year"] >= start_y) & (clustered_df["Year"] <= end_y)
        chunk = clustered_df[mask].copy()
        if chunk.empty:
            continue

        # For each cluster in this 3-year window
        for c_id in sorted(chunk["KMeans_Cluster"].unique()):
            df_cluster = chunk[chunk["KMeans_Cluster"] == c_id]
            result = train_models_on_cluster(
                df_cluster=df_cluster,
                market_label=market_label,
                start_yr=start_y,
                end_yr=end_y,
                cluster_id=c_id,
                feature_cols=feature_cols,
                target_col="VALUATION_VALUE",
                nn_epochs=nn_epochs,
                nn_patience=nn_patience,
                nn_models_dir="nn_models"
            )
            if result is not None:
                model_dict[((start_y, end_y), c_id)] = result

    return model_dict

###############################################################################
#  MAIN SCRIPT EXAMPLE:
###############################################################################
#   We'll assume you already have:
#       df_pierw_clustered_3yr, df_wtorn_clustered_3yr, df_nieokr_clustered_3yr
#       and also the KMeans dictionaries: pierw_kmeans_models, etc.

print("\n[INFO] Training LR & NN per cluster for Primary Market (pierw)...")
pierw_models_dict = train_models_for_all_clusters(
    df_pierw_clustered_3yr,
    "pierw",
    feature_cols = [
    "PREMISSES_USABLE_AREA",
    "PREMISSES_TECHNICAL_CONDITION",
    "CONSTRUCTION_YEAR",
    "PREMISSES_CHAMBER_NO",
    "BUILDING_TECHNICAL_CONDITION"
]
,
    nn_epochs=2000,
    nn_patience=500
)

print("\n[INFO] Training LR & NN per cluster for Secondary Market (wtorn)...")
wtorn_models_dict = train_models_for_all_clusters(
    df_wtorn_clustered_3yr,
    "wtorn",
    nn_epochs=2000,
    nn_patience=500
)

print("\n[INFO] Training LR & NN per cluster for Non-Commercial (nieokr)...")
nieokr_models_dict = train_models_for_all_clusters(
    df_nieokr_clustered_3yr,
    "nieokr",
    nn_epochs=2000,
    nn_patience=500
)

# Suppose you already have:
#   pierw_kmeans_models, wtorn_kmeans_models, nieokr_kmeans_models
# from the Part 1 KMeans clustering code.

all_market_models = {
    "pierw": {
        "kmeans": pierw_models,       # replaced pierw_kmeans_models with pierw_models
        "regressors": pierw_models_dict
    },
    "wtorn": {
        "kmeans": wtorn_models,       # replaced wtorn_kmeans_models with wtorn_models
        "regressors": wtorn_models_dict
    },
    "nieokr": {
        "kmeans": nieokr_models,      # replaced nieokr_kmeans_models with nieokr_models
        "regressors": nieokr_models_dict
    }
}

# Finally, save everything except the NN state_dicts themselves,
# because each cluster's NN weights are already in nn_models/*.pth
joblib.dump(all_market_models, "all_market_models.pkl")
print("\n[INFO] Saved all LR/NN references and KMeans objects to 'all_market_models.pkl'.")
print("[INFO] Individual .pth NN files are in the 'nn_models/' folder.")
```

```python
import os
import numpy as np
import pandas as pd
import joblib
import torch
import torch.nn as nn

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

###############################################################################
# (1) LOAD THE TRAINED MODELS FROM DISK
###############################################################################
# This file should have been created after your training script:
# It contains a dictionary of the form:
#   {
#     "pierw": {
#         "kmeans": {(start_yr,end_yr): (scaler, kmeans_model, k), ...},
#         "regressors": {((start_yr,end_yr), cluster_id): {
#             "lr_model": ...,
#             "nn_path": "...",
#             "nn_scaler": ...,
#             "nn_input_dim": ...,
#             "feature_cols": [...]
#         }, ...}
#     },
#     "wtorn": { ... },
#     "nieokr": { ... }
#   }
all_market_models = joblib.load("all_market_models.pkl")
print("[INFO] Loaded 'all_market_models.pkl' with KMeans + LR/NN references.")


###############################################################################
# (2) DEFINE A MATCHING PyTorch MODEL CLASS FOR LOADING STATE_DICT
###############################################################################
class FeedForwardNN(nn.Module):
    def __init__(self, input_dim, final_activation='prelu'):
        super(FeedForwardNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.first_activation = nn.LeakyReLU(0.1)

        if final_activation.lower() == 'leaky_relu':
            self.final_activation = nn.LeakyReLU(0.1)
        elif final_activation.lower() == 'prelu':
            self.final_activation = nn.PReLU()
        elif final_activation.lower() == 'gelu':
            self.final_activation = nn.GELU()
        elif final_activation.lower() == 'softplus':
            self.final_activation = nn.Softplus()
        else:
            self.final_activation = nn.ReLU()

        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        x = self.fc1(x)
        x = self.first_activation(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.dropout(x)
        x = self.fc3(x)
        x = self.final_activation(x)
        return x


###############################################################################
# (3) READ & CLEAN THE EVALUATION DATA (Main_task_data_evaluation.csv)
###############################################################################
df_eval = pd.read_csv(
    "Main_task_data_evaluation.csv",
    encoding="ISO-8859-1",
    sep=";",
    on_bad_lines="skip"
)

# Drop the same columns
columns_to_drop = [
    "SUBTYPE", "SHARE", "RENEWABLE_ENERGY_HEATING", "RENEWABLE_ENERGY_ELECTRIC",
    "CERTIFICATE_PHI", "COMMUNITY", "UTIL_WATER_INTAKE", "UTIL_WATER_SUPPLY",
    "UTIL_ELECTRICITY", "UTIL_GAS", "UTIL_SEWAGE_SYSTEM_CONNECTION",
    "UTIL_SEWAGE_SYSTEM_OWN", "UTIL_SEWAGE_SYSTEM_CONNECTION", "PREMISSES_STANDARD",
    "PREMISSES_INDEX_PED", "PREMISSES_INDEX_FED", "PREMISSES_INDEX_UED",
    "PREMISSES_ENERGY_PERF_CERT_DATE", "PREMISSES_ENERGY_PERF_CERT_VALI",
    "PREMISSES_RES_SHARE", "PREMISSES_CO2_EMMISSION", "BUILDING_CHAMBER_NO",
    "BUILDING_INDEX_PED", "BUILDING_INDEX_FED", "BUILDING_INDEX_UED",
    "BUILDING_ENERGY_PERF_CERT_DATE", "BUILDING_ENERGY_PERF_CERT_VALI",
    "BUILDING_RES_SHARE", "BUILDING_CO2_EMMISSION", "PARKING_SPACE_ID",
    "PARKING_KIND", "NUMBER"
]
df_eval.drop(columns=columns_to_drop, inplace=True, errors='ignore')

# Re-encode object columns to UTF-8
for col in df_eval.select_dtypes(include="object").columns:
    df_eval[col] = df_eval[col].apply(
        lambda x: x.encode("ISO-8859-1", errors="replace").decode("utf-8", errors="replace")
        if isinstance(x, str) else x
    )

# If "VALUATION_VALUE" is missing or you want to set it to 0, do so:
if "VALUATION_VALUE" not in df_eval.columns:
    df_eval["VALUATION_VALUE"] = 0
else:
    # Convert to numeric, fill missing with 0 if desired
    df_eval["VALUATION_VALUE"] = pd.to_numeric(df_eval["VALUATION_VALUE"], errors="coerce").fillna(0)

# City / capital cleaning
voivodeship_capitals = [
    "Warszawa", "Kraków", "Łódź", "Wrocław", "Poznań", "Gdańsk", "Szczecin",
    "Bydgoszcz", "Lublin", "Białystok", "Katowice", "Kielce", "Olsztyn",
    "Rzeszów", "Opole", "Zielona Góra"
]
df_eval["CITY_CLEAN"] = df_eval["CITY"].astype(str).str.strip().str.title()
df_eval["IS_CAPITAL"] = df_eval["CITY_CLEAN"].apply(lambda x: 1 if x in voivodeship_capitals else 0)

# Date fixes
df_eval["VALUE_DATE"] = pd.to_datetime(df_eval["VALUE_DATE"], errors="coerce")
max_date = pd.to_datetime("2025-04-02")
df_eval.loc[df_eval["VALUE_DATE"] > max_date, "VALUE_DATE"] = max_date
df_eval["VALUE_DATE"] = df_eval["VALUE_DATE"].fillna(max_date)
df_eval["VALUE_DATE_quarter"] = df_eval["VALUE_DATE"].dt.to_period("Q").astype(str)

# Label encoding for new columns if needed
# (In practice, you should do the same transforms as training.
#  If you have saved your LabelEncoders, re-load them. Otherwise do the same logic.)
exclude_cols = ["CITY_CLEAN", "VALUE_DATE_quarter"]
for col in df_eval.select_dtypes(include=["object"]).columns:
    if col not in exclude_cols:
        le = LabelEncoder()
        df_eval[col] = le.fit_transform(df_eval[col].astype(str))

# Numeric cleaning
df_eval["CONSTRUCTION_YEAR"] = pd.to_numeric(df_eval["CONSTRUCTION_YEAR"], errors="coerce")
valid_mean_year = df_eval.loc[df_eval["CONSTRUCTION_YEAR"] >= 500, "CONSTRUCTION_YEAR"].mean()
df_eval.loc[df_eval["CONSTRUCTION_YEAR"] < 500, "CONSTRUCTION_YEAR"] = valid_mean_year
df_eval["CONSTRUCTION_YEAR"] = df_eval["CONSTRUCTION_YEAR"].fillna(valid_mean_year)

df_eval["PREMISSES_USABLE_AREA"] = pd.to_numeric(df_eval["PREMISSES_USABLE_AREA"], errors="coerce")
threshold = df_eval["PREMISSES_USABLE_AREA"].quantile(0.99)
valid_mean_area = df_eval.loc[df_eval["PREMISSES_USABLE_AREA"] <= threshold, "PREMISSES_USABLE_AREA"].mean()
df_eval.loc[df_eval["PREMISSES_USABLE_AREA"] > threshold, "PREMISSES_USABLE_AREA"] = valid_mean_area
df_eval["PREMISSES_USABLE_AREA"] = df_eval["PREMISSES_USABLE_AREA"].fillna(valid_mean_area)

df_eval["STOREY"] = pd.to_numeric(df_eval["STOREY"], errors="coerce").fillna(df_eval["STOREY"].mean())
df_eval["PARCEL_AREA"] = pd.to_numeric(df_eval["PARCEL_AREA"], errors="coerce").fillna(df_eval["PARCEL_AREA"].mean())
df_eval["FLOORS_NO"] = pd.to_numeric(df_eval["FLOORS_NO"], errors="coerce").fillna(df_eval["FLOORS_NO"].mean())

df_eval.dropna(inplace=True)

# Create "Year" from "VALUE_DATE_quarter"
def parse_year_quarter(qstr):
    return int(qstr[:4])

df_eval["Year"] = df_eval["VALUE_DATE_quarter"].apply(parse_year_quarter)

# We'll assume "MARKET_TYPE" is in the data (1=pierw,2=wtorn,0=nieokr).
# If it's missing or coded differently, adjust accordingly.
df_eval_pierw = df_eval[df_eval["MARKET_TYPE"] == 1].copy()
df_eval_wtorn = df_eval[df_eval["MARKET_TYPE"] == 2].copy()
df_eval_nieokr = df_eval[df_eval["MARKET_TYPE"] == 0].copy()

df_eval_pierw["Market_Type"] = "pierw"
df_eval_wtorn["Market_Type"] = "wtorn"
df_eval_nieokr["Market_Type"] = "nieokr"

# We'll create columns for results
df_eval["Assigned_Cluster"] = -1
df_eval["Pred_LR"] = np.nan
df_eval["Pred_NN"] = np.nan


###############################################################################
# (4) HELPER FUNCTIONS TO ASSIGN CLUSTERS & PREDICT
###############################################################################
def get_kmeans_model(market_type, start_yr, end_yr):
    """
    From all_market_models[market_type]["kmeans"], get the (scaler, kmeans, k).
    """
    kmeans_dict = all_market_models[market_type]["kmeans"]  # => {(start,end): (scaler,kmeans,k)}
    return kmeans_dict[(start_yr, end_yr)]  # => (scaler, kmeans_obj, k)

def get_regressors(market_type):
    """
    Return the dict of cluster models for the given market type:
      { ((start_yr, end_yr), cluster_id): { "lr_model":..., "nn_path":..., etc.} }
    """
    return all_market_models[market_type]["regressors"]

def load_nn_model(nn_path, input_dim, final_activation="prelu"):
    """
    Build a new FeedForwardNN, load its state_dict from nn_path, return it.
    """
    model = FeedForwardNN(input_dim=input_dim, final_activation=final_activation)
    model.load_state_dict(torch.load(nn_path))
    model.eval()
    return model

def assign_clusters_and_predict(df_subset, market_label):
    """
    1) For each 3-year window in the saved KMeans dictionary, find the matching rows.
    2) Use (scaler, kmeans) to assign cluster.
    3) Then load the LR & NN for that cluster, generate predictions, store in df_subset.
    """
    kmeans_dict = all_market_models[market_label]["kmeans"]     # => {(start_yr,end_yr): (scaler,kmeans,k),...}
    regressors_dict = all_market_models[market_label]["regressors"]

    # We'll define a set of base cluster features for assignment
    cluster_feats = ["PREMISSES_USABLE_AREA", "PREMISSES_TECHNICAL_CONDITION", "VALUATION_VALUE", "CONSTRUCTION_YEAR"]

    # For each 3-year window:
    for (start_yr, end_yr), (km_scaler, km_model, k_opt) in kmeans_dict.items():
        # mask the rows
        mask = (df_subset["Year"] >= start_yr) & (df_subset["Year"] <= end_yr)
        chunk = df_subset.loc[mask]
        if chunk.empty:
            continue

        X_chunk = chunk[cluster_feats].values
        X_scaled = km_scaler.transform(X_chunk)
        clusters_assigned = km_model.predict(X_scaled)

        # Save assigned cluster
        df_subset.loc[mask, "Assigned_Cluster"] = clusters_assigned

        # Now for each cluster_id in that chunk, run LR/NN
        for c_id in np.unique(clusters_assigned):
            # mask for cluster c_id
            c_mask = (df_subset.index.isin(chunk.index)) & (df_subset["Assigned_Cluster"] == c_id)
            # Look up the LR/NN dictionary
            reg_key = ((start_yr, end_yr), c_id)
            if reg_key not in regressors_dict:
                # This cluster might not exist or was skipped in training
                continue
            reg_info = regressors_dict[reg_key]

            # Extract the LR model, nn_path, nn_scaler, etc.
            lr_model   = reg_info["lr_model"]
            nn_path    = reg_info["nn_path"]
            nn_scaler  = reg_info["nn_scaler"]
            nn_input   = reg_info["nn_input_dim"]
            feat_cols  = reg_info["feature_cols"]  # e.g. ["PREMISSES_USABLE_AREA", "PREMISSES_TECHNICAL_CONDITION", "CONSTRUCTION_YEAR", "STOREY"]

            # Prepare the subset's features for LR
            X_feat_LR = df_subset.loc[c_mask, feat_cols].values
            # Predict LR
            lr_pred = lr_model.predict(X_feat_LR)
            df_subset.loc[c_mask, "Pred_LR"] = lr_pred

            # Prepare for NN
            X_feat_nn = nn_scaler.transform(X_feat_LR)
            X_torch   = torch.tensor(X_feat_nn, dtype=torch.float32)

            # Load the NN model
            nn_model = load_nn_model(nn_path, nn_input, final_activation="prelu")
            with torch.no_grad():
                pred_nn = nn_model(X_torch).numpy().flatten()

            df_subset.loc[c_mask, "Pred_NN"] = pred_nn

    return df_subset


###############################################################################
# (5) APPLY THE FUNCTIONS TO PIERW, WTORN, NIEOKR EVAL SUBSETS
###############################################################################
df_eval_pierw = assign_clusters_and_predict(df_eval_pierw, "pierw")
df_eval_wtorn = assign_clusters_and_predict(df_eval_wtorn, "wtorn")
df_eval_nieokr = assign_clusters_and_predict(df_eval_nieokr, "nieokr")

# Combine back
df_eval_combined = pd.concat([df_eval_pierw, df_eval_wtorn, df_eval_nieokr], axis=0)

# For convenience, sort by original index if you want:
df_eval_combined = df_eval_combined.sort_index()

# Now "df_eval_combined" has columns: Assigned_Cluster, Pred_LR, Pred_NN, etc.
# If you want to save them into the original CSV, do so:

df_eval_combined.to_csv("Main_task_data_evaluation_with_predictions.csv", sep=";", encoding="utf-8-sig", index=False)
print("[INFO] Saved predictions to 'Main_task_data_evaluation_with_predictions.csv'.")
```

```python
import os
import numpy as np
import pandas as pd
import joblib
import torch
import torch.nn as nn

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

###############################################################################
# (1) LOAD THE TRAINED MODELS FROM DISK
###############################################################################
all_market_models = joblib.load("all_market_models.pkl")
print("[INFO] Loaded 'all_market_models.pkl' with KMeans + LR/NN references.")


###############################################################################
# (2) DEFINE A MATCHING PyTorch MODEL CLASS FOR LOADING STATE_DICT
###############################################################################
class FeedForwardNN(nn.Module):
    def __init__(self, input_dim, final_activation='prelu'):
        super(FeedForwardNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.first_activation = nn.LeakyReLU(0.1)

        if final_activation.lower() == 'leaky_relu':
            self.final_activation = nn.LeakyReLU(0.1)
        elif final_activation.lower() == 'prelu':
            self.final_activation = nn.PReLU()
        elif final_activation.lower() == 'gelu':
            self.final_activation = nn.GELU()
        elif final_activation.lower() == 'softplus':
            self.final_activation = nn.Softplus()
        else:
            self.final_activation = nn.ReLU()

        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        x = self.fc1(x)
        x = self.first_activation(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.dropout(x)
        x = self.fc3(x)
        x = self.final_activation(x)
        return x


###############################################################################
# (3) READ & CLEAN THE EVALUATION DATA (Main_task_data_evaluation.csv)
###############################################################################
df_eval = pd.read_csv(
    "Main_task_data_evaluation.csv",
    encoding="ISO-8859-1",
    sep=";",
    on_bad_lines="skip"
)

# Drop the same columns
columns_to_drop = [
    "SUBTYPE", "SHARE", "RENEWABLE_ENERGY_HEATING", "RENEWABLE_ENERGY_ELECTRIC",
    "CERTIFICATE_PHI", "COMMUNITY", "UTIL_WATER_INTAKE", "UTIL_WATER_SUPPLY",
    "UTIL_ELECTRICITY", "UTIL_GAS", "UTIL_SEWAGE_SYSTEM_CONNECTION",
    "UTIL_SEWAGE_SYSTEM_OWN", "UTIL_SEWAGE_SYSTEM_CONNECTION", "PREMISSES_STANDARD",
    "PREMISSES_INDEX_PED", "PREMISSES_INDEX_FED", "PREMISSES_INDEX_UED",
    "PREMISSES_ENERGY_PERF_CERT_DATE", "PREMISSES_ENERGY_PERF_CERT_VALI",
    "PREMISSES_RES_SHARE", "PREMISSES_CO2_EMMISSION", "BUILDING_CHAMBER_NO",
    "BUILDING_INDEX_PED", "BUILDING_INDEX_FED", "BUILDING_INDEX_UED",
    "BUILDING_ENERGY_PERF_CERT_DATE", "BUILDING_ENERGY_PERF_CERT_VALI",
    "BUILDING_RES_SHARE", "BUILDING_CO2_EMMISSION", "PARKING_SPACE_ID",
    "PARKING_KIND", "NUMBER"
]
df_eval.drop(columns=columns_to_drop, inplace=True, errors='ignore')

# Re-encode object columns to UTF-8
for col in df_eval.select_dtypes(include="object").columns:
    df_eval[col] = df_eval[col].apply(
        lambda x: x.encode("ISO-8859-1", errors="replace").decode("utf-8", errors="replace")
        if isinstance(x, str) else x
    )

# If "VALUATION_VALUE" is missing or you want to set it to 0, do so:
if "VALUATION_VALUE" not in df_eval.columns:
    df_eval["VALUATION_VALUE"] = 0
else:
    df_eval["VALUATION_VALUE"] = pd.to_numeric(df_eval["VALUATION_VALUE"], errors="coerce").fillna(0)

voivodeship_capitals = [
    "Warszawa", "Kraków", "Łódź", "Wrocław", "Poznań", "Gdańsk", "Szczecin",
    "Bydgoszcz", "Lublin", "Białystok", "Katowice", "Kielce", "Olsztyn",
    "Rzeszów", "Opole", "Zielona Góra"
]
df_eval["CITY_CLEAN"] = df_eval["CITY"].astype(str).str.strip().str.title()
df_eval["IS_CAPITAL"] = df_eval["CITY_CLEAN"].apply(lambda x: 1 if x in voivodeship_capitals else 0)

df_eval["VALUE_DATE"] = pd.to_datetime(df_eval["VALUE_DATE"], errors="coerce")
max_date = pd.to_datetime("2025-04-02")
df_eval.loc[df_eval["VALUE_DATE"] > max_date, "VALUE_DATE"] = max_date
df_eval["VALUE_DATE"] = df_eval["VALUE_DATE"].fillna(max_date)
df_eval["VALUE_DATE_quarter"] = df_eval["VALUE_DATE"].dt.to_period("Q").astype(str)

exclude_cols = ["CITY_CLEAN", "VALUE_DATE_quarter"]
for col in df_eval.select_dtypes(include=["object"]).columns:
    if col not in exclude_cols:
        le = LabelEncoder()
        df_eval[col] = le.fit_transform(df_eval[col].astype(str))

df_eval["CONSTRUCTION_YEAR"] = pd.to_numeric(df_eval["CONSTRUCTION_YEAR"], errors="coerce")
valid_mean_year = df_eval.loc[df_eval["CONSTRUCTION_YEAR"] >= 500, "CONSTRUCTION_YEAR"].mean()
df_eval.loc[df_eval["CONSTRUCTION_YEAR"] < 500, "CONSTRUCTION_YEAR"] = valid_mean_year
df_eval["CONSTRUCTION_YEAR"] = df_eval["CONSTRUCTION_YEAR"].fillna(valid_mean_year)

df_eval["PREMISSES_USABLE_AREA"] = pd.to_numeric(df_eval["PREMISSES_USABLE_AREA"], errors="coerce")
threshold = df_eval["PREMISSES_USABLE_AREA"].quantile(0.99)
valid_mean_area = df_eval.loc[df_eval["PREMISSES_USABLE_AREA"] <= threshold, "PREMISSES_USABLE_AREA"].mean()
df_eval.loc[df_eval["PREMISSES_USABLE_AREA"] > threshold, "PREMISSES_USABLE_AREA"] = valid_mean_area
df_eval["PREMISSES_USABLE_AREA"] = df_eval["PREMISSES_USABLE_AREA"].fillna(valid_mean_area)

df_eval["STOREY"] = pd.to_numeric(df_eval["STOREY"], errors="coerce").fillna(df_eval["STOREY"].mean())
df_eval["PARCEL_AREA"] = pd.to_numeric(df_eval["PARCEL_AREA"], errors="coerce").fillna(df_eval["PARCEL_AREA"].mean())
df_eval["FLOORS_NO"] = pd.to_numeric(df_eval["FLOORS_NO"], errors="coerce").fillna(df_eval["FLOORS_NO"].mean())

df_eval.dropna(inplace=True)

def parse_year_quarter(qstr):
    return int(qstr[:4])

df_eval["Year"] = df_eval["VALUE_DATE_quarter"].apply(parse_year_quarter)

# Suppose 1= 'pierw', 2= 'wtorn', 0='nieokr'
df_eval_pierw = df_eval[df_eval["MARKET_TYPE"] == 1].copy()
df_eval_wtorn = df_eval[df_eval["MARKET_TYPE"] == 2].copy()
df_eval_nieokr = df_eval[df_eval["MARKET_TYPE"] == 0].copy()

df_eval_pierw["Market_Type"] = "pierw"
df_eval_wtorn["Market_Type"] = "wtorn"
df_eval_nieokr["Market_Type"] = "nieokr"

df_eval["Assigned_Cluster"] = -1
df_eval["Pred_LR"] = np.nan
df_eval["Pred_NN"] = np.nan


###############################################################################
# (4) HELPER FUNCTIONS TO ASSIGN CLUSTERS & PREDICT
###############################################################################
def get_kmeans_model(market_type, start_yr, end_yr):
    kmeans_dict = all_market_models[market_type]["kmeans"]
    return kmeans_dict[(start_yr, end_yr)]

def get_regressors(market_type):
    return all_market_models[market_type]["regressors"]

def load_nn_model(nn_path, input_dim, final_activation="prelu"):
    model = FeedForwardNN(input_dim=input_dim, final_activation=final_activation)
    model.load_state_dict(torch.load(nn_path))
    model.eval()
    return model

def assign_clusters_and_predict(df_subset, market_label):
    # For cluster assignment, we do NOT change anything:
    cluster_feats = [
        "PREMISSES_USABLE_AREA",
        "PREMISSES_TECHNICAL_CONDITION",
        "VALUATION_VALUE",
        "CONSTRUCTION_YEAR"
    ]
    kmeans_dict = all_market_models[market_label]["kmeans"]
    regressors_dict = all_market_models[market_label]["regressors"]

    for (start_yr, end_yr), (km_scaler, km_model, k_opt) in kmeans_dict.items():
        mask = (df_subset["Year"] >= start_yr) & (df_subset["Year"] <= end_yr)
        chunk = df_subset.loc[mask]
        if chunk.empty:
            continue

        X_chunk = chunk[cluster_feats].values
        X_scaled = km_scaler.transform(X_chunk)
        clusters_assigned = km_model.predict(X_scaled)

        df_subset.loc[mask, "Assigned_Cluster"] = clusters_assigned

        for c_id in np.unique(clusters_assigned):
            c_mask = (df_subset.index.isin(chunk.index)) & (df_subset["Assigned_Cluster"] == c_id)
            reg_key = ((start_yr, end_yr), c_id)
            if reg_key not in regressors_dict:
                continue

            # Retrieve the references used by the NN
            reg_info = regressors_dict[reg_key]
            lr_model   = reg_info["lr_model"]
            nn_path    = reg_info["nn_path"]
            nn_scaler  = reg_info["nn_scaler"]
            nn_input   = reg_info["nn_input_dim"]

            # Here is the 5-column set used by the NN:
            feat_cols  = reg_info["feature_cols"]
            # e.g. ["PREMISSES_USABLE_AREA","PREMISSES_TECHNICAL_CONDITION",
            #      "CONSTRUCTION_YEAR","PREMISSES_CHAMBER_NO",
            #      "BUILDING_TECHNICAL_CONDITION"]

            print(f"[INFO] For cluster=({start_yr},{end_yr}), c_id={c_id}, the NN training features were: {feat_cols}")

            # 1) LR Predictions
            X_feat_LR = df_subset.loc[c_mask, feat_cols].values
            lr_pred = lr_model.predict(X_feat_LR)
            df_subset.loc[c_mask, "Pred_LR"] = lr_pred

            # 2) NN Predictions
            X_feat_nn = nn_scaler.transform(X_feat_LR)
            X_torch   = torch.tensor(X_feat_nn, dtype=torch.float32)
            nn_model = load_nn_model(nn_path, nn_input, final_activation="prelu")
            with torch.no_grad():
                pred_nn = nn_model(X_torch).numpy().flatten()

            df_subset.loc[c_mask, "Pred_NN"] = pred_nn

    return df_subset


###############################################################################
# (5) APPLY THE FUNCTIONS TO PIERW, WTORN, NIEOKR EVAL SUBSETS
###############################################################################
df_eval_pierw = assign_clusters_and_predict(df_eval_pierw, "pierw")
df_eval_wtorn = assign_clusters_and_predict(df_eval_wtorn, "wtorn")
df_eval_nieokr = assign_clusters_and_predict(df_eval_nieokr, "nieokr")

df_eval_combined = pd.concat([df_eval_pierw, df_eval_wtorn, df_eval_nieokr], axis=0).sort_index()

df_eval_combined.to_csv("Main_task_data_evaluation_with_predictions.csv", sep=";", encoding="utf-8-sig", index=False)
print("[INFO] Predictions saved to 'Main_task_data_evaluation_with_predictions.csv'.")


# ------------------- (OPTIONAL) SHAP EXPLANATIONS -------------------
import shap

market_label_example = "pierw"
year_range_example = (2015, 2017)
cluster_id_example = 0

regressors_dict_example = all_market_models[market_label_example]["regressors"]
reg_key = (year_range_example, cluster_id_example)

if reg_key in regressors_dict_example:
    reg_info_example = regressors_dict_example[reg_key]
    lr_model_example = reg_info_example["lr_model"]
    feat_cols_example = reg_info_example["feature_cols"]  # the 5 columns

    sample_mask = (
        (df_eval_combined["Market_Type"] == market_label_example) &
        (df_eval_combined["Assigned_Cluster"] == cluster_id_example) &
        (df_eval_combined["Year"] >= year_range_example[0]) &
        (df_eval_combined["Year"] <= year_range_example[1])
    )
    df_sample = df_eval_combined.loc[sample_mask]

    if not df_sample.empty:
        X_sample = df_sample[feat_cols_example].values

        explainer = shap.Explainer(lr_model_example, X_sample, feature_names=feat_cols_example)
        shap_values = explainer(X_sample)

        print("[INFO] SHAP summary for cluster_id=0 in pierw (2015-2017):")

        # 1) Bar Plot
        shap.plots.bar(shap_values)

        # 2) Beeswarm Plot
        shap.plots.beeswarm(shap_values)
    else:
        print(f"[WARNING] No rows in evaluation data for cluster_id=0 in range=(2015,2017).")
else:
    print("[WARNING] This cluster doesn't exist in 'pierw' for that year range.")
```
